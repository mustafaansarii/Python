```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data=sns.load_dataset("tips") #from seaborn github
print(data)
sns.barplot(data=data,x='day',y="tip",   estimator='median')
```

```python
sns.barplot(data=data,x='day',y="tip",   estimator='sum')
```

```python
sns.barplot(data=data,x='day',y="tip",  hue='sex',palette="spring")
```

```python
sns.barplot(data=data,x='day',y="tip",  hue='sex',palette="spring",order=["Sun","Thur","Fri","Sat"])
```

```python
sns.barplot(data=data,x='day',y="tip",  hue='sex',palette="spring",order=["Sun","Thur","Fri","Sat"],errorbar=("ci",0))
```
