```python
import seaborn as sns
import pandas as pd
data={"days":[1,2,3,4,5],
     "NOP":[32,34,56,78,90]}
df=pd.DataFrame(data)
df
sns.lineplot(data=data,x="days",y="NOP")
```

```python
data=pd.read_excel("ESD.xlsx")
df=pd.DataFrame(data)
df
color=sns.color_palette("viridis")

sns.lineplot(data=df,x="Business Unit",y="Age",hue="Gender",palette=color)
```

```python
color=sns.color_palette("GnBu")
sns.lineplot(data=data,x="Business Unit",y="Age",hue="Gender",style="Ethnicity")
```
