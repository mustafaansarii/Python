A count plot is a type of categorical plot in Seaborn that displays the count of observations in each category using bars. It's useful for visualizing the distribution of a categorical variable and understanding the frequency of each category in the dataset. The sns.countplot function in Seaborn is typically used to create count plots.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data=sns.load_dataset("tips")
data
```

```python
data=data.head(10)
sns.countplot(data,x="total_bill")
```

```python
sns.countplot(data,x="sex")
```

```python
sns.countplot(data,x="day")
```

```python
data=pd.read_excel("ESD.xlsx")
data=data.head(20)
data
```

```python
sns.countplot(data,x="Department",hue="Gender",palette="viridis")
plt.show()
```

```python
sns.countplot(data,x="Business Unit",hue="Gender",palette="GnBu")
plt.show()
```
