
A pair plot in Seaborn is a grid of scatterplots and histograms that allows you to visualize the relationships between pairs of variables in a dataset. It's a convenient way to explore the joint distribution of numerical features and to identify patterns, correlations, and potential outliers. The sns.pairplot function is commonly used to create pair plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_excel("ESD.xlsx")
data=data.head(20)
data
```

```python
sns.pairplot(data)
```

```python
data=sns.load_dataset("tips")
sns.pairplot(data,hue="day",diag_kind="kde")
```

```python
data=sns.load_dataset("iris")
data
```

```python
sns.pairplot(data,hue="species",diag_kind="kde")
```

```python
sns.pairplot(data,hue="species",diag_kind="hist")
```
