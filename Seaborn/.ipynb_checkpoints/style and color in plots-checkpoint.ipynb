```python
import seaborn as sns
import matplotlib.pyplot as plt
data=sns.load_dataset("exercise")
data=data.head(25)
```

```python
sns.barplot(data,x="pulse",y="id",hue='diet')
```

```python
sns.set_style(style="dark")
sns.barplot(data,x="pulse",y="id",hue='diet',)
```

```python
sns.set_style(style="dark")
sns.barplot(data,x="pulse",y="id",hue='diet',)
```

```python
sns.set_style(style="whitegrid")
sns.barplot(data,x="pulse",y="id",hue='diet',)
```

```python
sns.set_style(style="ticks")
sns.barplot(data,x="pulse",y="id",hue='diet',)
```
