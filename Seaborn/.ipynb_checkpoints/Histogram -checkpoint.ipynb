```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset
tips = sns.load_dataset("tips")
tips


```

```python
# Bar plot for categorical data
sns.barplot(x="day", y="total_bill", data=tips, hue="sex", ci=None, palette="Set2")
plt.title("Bar Plot")
plt.show()
```

```python

# Heatmap to visualize the correlation matrix of numerical features
correlation_matrix = tips.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Violin plot for a combination of box plot and kernel density estimation
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True, palette="pastel")
plt.title("Violin Plot")
plt.show()

```

```python
# Box plot to visualize distribution and outliers
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex", palette="coolwarm")
plt.title("Box Plot")
plt.show()
```

```python

# Pair plot for multiple numerical features
sns.pairplot(tips, hue="sex", palette="husl")
plt.title("Pair Plot")
plt.show()

```

```python
# Scatter plot with regression line
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot with Regression Line")
plt.show()
```
