```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=sns.load_dataset("tips")
data
gp=data.groupby("day").agg({"tip":"mean"})
gp
```

```python
sns.heatmap(gp)
```
