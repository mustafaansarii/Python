```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
data=sns.load_dataset("tips")
data
```

```python

sns.scatterplot(data,x='total_bill',y='tip')
```

```python
sns.scatterplot(data,x='total_bill',y='tip',hue='smoker')
```

```python
sns.scatterplot(data,x='total_bill',y='tip',hue='day')
```

```python
sns.scatterplot(data,x='total_bill',y='tip',hue='day',size='size')
```

```python
sns.scatterplot(data,x='total_bill',y='tip',hue='day',marker="d")
```

```python
sns.scatterplot(data,x='total_bill',y='tip',hue='day',palette="spring")
```

```python
sns.scatterplot(data,x='total_bill',y='tip',hue='day',palette="viridis")
```
