```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
tips = sns.load_dataset("tips")
sns.kdeplot(tips,x="total_bill")
```

```python
sns.histplot(tips,x="total_bill")
```

```python
sns.kdeplot(tips,x="total_bill",hue="day")
```

```python
sns.kdeplot(tips,x="total_bill",hue='day',multiple="stack")
```

```python
sns.kdeplot(tips,x="total_bill",hue='day',multiple="fill")
```

```python
sns.kdeplot(tips,x="total_bill",hue='day',multiple="fill",color="skyblue")
```
