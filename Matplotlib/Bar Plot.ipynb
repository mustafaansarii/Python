```python
import numpy as np
import matplotlib.pyplot as plt

# Generating sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating a simple line plot
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-', marker='o')
plt.plot(x, y2, label='cos(x)', color='green', linestyle='--', marker='x')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sin(x) and Cos(x)')

# Adding a legend
plt.legend()

# Setting axis limits
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

# Adding grid
plt.grid(True)

# Adding text annotation
plt.text(2, 1, 'Text Annotation', fontsize=10, color='red')

# Adding vertical and horizontal lines
plt.axvline(x=5, color='gray', linestyle='--', linewidth=2)
plt.axhline(y=0, color='black', linestyle='-', linewidth=2)

# Adding subplots
plt.subplot(2, 1, 2)
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sin(x) in Subplot')

# Saving the figure
plt.savefig('example_plot.png')

# Displaying the plot
plt.show()
```
