```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

day = [1, 2, 3, 4, 5, 6]
nop1 = [50, 4, 23, 45, 67, 89]
nop2 = [12, 23, 4, 53, 96, 97]
nop3 = [23, 56, 76, 87, 90, 99]

plt.stackplot(day, nop1, nop2, nop3, labels=['nop1', 'nop2', 'nop3'])
plt.legend(loc='upper left')
plt.xlabel('Day')
plt.ylabel('Values')
plt.title('Stacked Plot Example')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already loaded the data into a DataFrame named 'df'
# If not, you can load it using: df = pd.read_excel('your_file.xlsx')

# Extract relevant columns
days = df.index  # Assuming the index represents the days
single_day_data = df['Single day data']
weekly_results_data = df['Weekly Results Data']

# Create a stack plot
plt.figure(figsize=(10, 6))
plt.stackplot(days, single_day_data, weekly_results_data, labels=['Single Day Data', 'Weekly Results Data'])
plt.legend(loc='upper left')
plt.title('Stack Plot of Single Day and Weekly Results Data')
plt.xlabel('Days')
plt.ylabel('Values')
plt.show()
```
