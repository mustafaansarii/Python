```python
import matplotlib.pyplot as plt
x=[1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
y=[543.56, 763.67, 897.56, 979.54, 1020.54, 1120.87, 1298.89, 1507.87]
plt.figure(figsize=(10, 8))
plt.plot(x,y, marker="o")
plt.xlabel("Year",fontsize=10)
plt.ylabel("$",fontsize=20)
plt.title("popularity of different parts of harry porter")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='blue', linestyle='-',marker="o")
```

```python
x=[1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
y=[543.56, 763.67, 897.56, 979.54, 1020.54, 1120.87, 1298.89, 1507.87]
plt.figure(figsize=(10, 8))
plt.bar(x,y)
plt.grid(True)
plt.xlabel("Year",fontsize=20)
plt.ylabel("$",fontsize=20)
```

```python
from collections import Counter
grades=[100,99,89,78,76,65,56,45,34,23,45,0]
histogram=Counter(min(grade//10*10,90)for grade in grades)
plt.figure(figsize=(10,6))
plt.bar([x+5 for x in histogram.keys()],histogram.values(),10,edgecolor="black")
plt.axis([-5,105,0,5])
plt.xticks([10* i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("# of student")
plt.title("Distribution Exam grades")
```
