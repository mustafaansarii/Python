```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('expense3.xlsx')
print(df.head())




```

```python
# Plot a pie chart of Payment Mode distribution
plt.figure(figsize=(8, 8))
payment_mode_counts = df['Payment Mode'].value_counts()
payment_mode_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightgreen', 'lightskyblue'])
plt.title('Payment Mode Distribution')
plt.show()
```

```python
# Plot a bar chart of total Amount for each Category
plt.figure(figsize=(10, 6))
category_amount = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
category_amount.plot(kind='bar', color='skyblue')
plt.title('Total Amount by Category')
plt.xlabel('Category')
plt.ylabel('Total Amount')
plt.show()
```
