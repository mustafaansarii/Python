```python

import pandas as pd
import matplotlib.pyplot as plt
x=["day1","day2","day3","day4","day5"]
y=[300,100,450,300,500]
y1=[400,450,455,350,400]
plt.plot(x,y,marker="o",color="black",ls="-",label="week2")
plt.plot(x,y1,marker="o",color="green",ls="-",label="week1",alpha=0.5)
plt.legend(loc=4)
plt.show()
```
