```python
import matplotlib.pyplot as plt
x=[12,22,23,24,25]
y=[22,33,44,55,66]
y1=[49,34,76,78,70]
y2=[1,36,54,55,87]
plt.figure(figsize=[5,3])
plt.plot(x,y,label="male")
plt.plot(x,y1,label="female")
plt.plot(x,y2,label="baby")
plt.plot(x,y2,label="young")
plt.legend(loc=0)
```

```python
plt.plot(x,y)
plt.plot(x,y1)
plt.legend(["a1","a2"],ncols=2)
```
