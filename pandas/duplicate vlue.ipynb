```python
import pandas as pd

# Reading data from an Excel file into a DataFrame
data = pd.read_excel("expense3.xlsx")

# Printing the DataFrame
print(data)
```

```python
print(data.duplicated())

# data.duplicated(): This Pandas method checks for duplicated rows in the DataFrame. It returns a boolean Series where each element indicates whether the corresponding row in the DataFrame is a duplicate or not.

# print(data.duplicated()): This prints the boolean Series to the console. If a row is a duplicate, the corresponding value in the Series will be True; otherwise, it will be False.
```

```python
# Checking for duplicate entries in the 'Name' column and printing the count
print(data['Name'].duplicated().sum())
```

```python
print(data['Name'].duplicated())
```

```python
print(data.drop_duplicates("Name")) #only uniq vlue
```
